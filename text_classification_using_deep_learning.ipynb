```python
col={x:x.lower() for x in df.columns}
df=df.rename(columns=col)
df.head(5)
```

```python
#lets bring the text in  ther colums into lower case letters
df['article']=df['article'].str.lower()
df['category']=df['category'].str.lower()
df.head(10)
```

```python
import re
concat_string=' '.join(df['article'].astype('str'))
concat_string=re.sub(r'[^A-Za-z0-9]', ' ',concat_string)
a={}
for i in concat_string.split(' '):
    if i not in a:
        a[i]=1
    else:
        a[i]+=1
df1=pd.DataFrame(columns=['word','freq'])
df1['word']=a.keys()
df1['freq']=a.values()
df1
```

```python
df2=df1.sort_values(by='freq',ascending=False)
df2.head(20)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download stopwords if not already
nltk.download('stopwords')
nltk.download('punkt')


# Get English stopwords
stop_words = set(stopwords.words('english'))

# Tokenize words
words = word_tokenize(concat_string)

# Remove stopwords
filtered = [word for word in words if word.lower() not in stop_words]

```

```python
df2=pd.DataFrame(columns=['word'],data=filtered)
df2
```

```python
df3=pd.merge(left=df2,right=df1,how='inner',on='word')
df3.head(2)
```

```python
df3=df3.drop_duplicates()
df3=df3.reset_index(drop=True)
df3
```
